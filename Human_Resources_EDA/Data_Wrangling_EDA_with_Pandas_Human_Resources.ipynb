# 8: Visualize Data: HISTOGRAM AND CORRELATION

- **Several features such as 'MonthlyIncome' and 'TotalWorkingYears' are tail heavy**
- **Job level is strongly correlated with total working hours**
- **Monthly income is strongly correlated with Job level**
- **Monthly income is strongly correlated with total working hours**
- **Age is strongly correlated with monthly income**

```python
hr_df = pd.read_csv('Human_Resources.csv')
hr_df.dropna(inplace=True)
hr_df.drop(['EmployeeCount', 'EmployeeNumber', 'StandardHours' , 'Over18'], axis=1, inplace=True)
hr_df
```

```python
hr_df.describe().T
```

```python
## numeric_tips = tips.select_dtypes(include=[np.number])
## correlation_matrix = numeric_tips.corr()
```

```python
hr_df.select_dtypes('number').corr()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
selected_columns = ['MonthlyIncome', 'TotalWorkingYears', 'JobLevel', 'Age']

selected_df = hr_df[selected_columns]

plt.figure(figsize=(10, 8))
plt.tight_layout()

# Heatmap
sns.heatmap(selected_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()

# Histogram
selected_df.hist(bins=20, figsize=(12, 10))
plt.suptitle('Histograms of Selected Features', x=0.5, y=1.02, ha='center', fontsize='large')
plt.show()
```
